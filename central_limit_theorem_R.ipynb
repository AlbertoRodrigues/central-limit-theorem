```r
require(ggplot2)
require(gridExtra)
install.packages("LaplacesDemon")
require(LaplacesDemon)
options(warn=-1)
```

```r
Sim.Pois = function(n,B,lambda){
  X  <- matrix(rpois(n*B,lambda),B,n)
  Zn <- sqrt(n)*(rowMeans(X)-lambda)/sqrt(lambda)
  (ggplot(data.frame(x=Zn),aes(x))+geom_histogram(aes(y =..density..),fill="#00AFBB",color="black")+
  stat_function(fun = dnorm, args = list( 0, 1),aes(fill="black"),size=1.2)+theme_minimal()+
      labs(title=substitute(paste("n= ",n,", B= ",B,", ",lambda,"= ",parametro,sep=""),
    list(n = match.call()$n,B=match.call()$B,parametro=match.call()$lambda))))
}
```

```r
a=Sim.Pois(50,100,2)
b=Sim.Pois(500,100,2)
c=Sim.Pois(500,500,2)
d=Sim.Pois(5000,2000,2)
e=Sim.Pois(500,200,0.01)
f=Sim.Pois(500,500,0.01)
g=Sim.Pois(300,2000,0.01)
h=Sim.Pois(5000,10000,0.01)
i=Sim.Pois(50,100,8.5)
j=Sim.Pois(500,100,8.5)
l=Sim.Pois(300,500,8.5)
m=Sim.Pois(500,5000,8.5)
```

```r
grid.arrange(a,b,c,d,ncol=2)
```

```r
grid.arrange(e,f,g,h,ncol=2)
```

```r
grid.arrange(i,j,l,m,ncol=2)
```

```r
Sim.Bern = function(n,B,p){
  X  <- matrix(rbinom(n*B,1,p),B,n)
  Zn <- sqrt(n)*(rowMeans(X)-p)/sqrt(p*(1-p))
  (ggplot(data.frame(x=Zn),aes(x))+geom_histogram(aes(y =..density..),fill="#00AFBB",color="black")+
      stat_function(fun = dnorm, args = list( 0, 1),aes(fill="black"),size=1.2)+theme_minimal()+
      labs(title=substitute(paste("n= ",n,", B= ",B,",p= ",parametro,sep=""),
                            list(n = match.call()$n,B=match.call()$B,parametro=match.call()$p))))
}
```

```r
a=Sim.Bern(500,100,0.01)
b=Sim.Bern(500,300,0.01)
c=Sim.Bern(500,2000,0.01)
d=Sim.Bern(10000,8000,0.01)
e=Sim.Bern(100,100,0.5)
f=Sim.Bern(500,300,0.5)
g=Sim.Bern(300,1000,0.5)
h=Sim.Bern(5000,2000,0.5)
i=Sim.Bern(300,100,0.95)
j=Sim.Bern(500,300,0.95)
l=Sim.Bern(500,2000,0.95)
m=Sim.Bern(1000,5000,0.95)
```

```r
Sim.Beta = function(n,B,a,b){
  X  <- matrix(rbeta(n*B,a,b),B,n)
  Zn <- sqrt(n)*(rowMeans(X)-(a/(a+b)))/sqrt((a*b)/((a+b+1)*(a+b)^2))
  (ggplot(data.frame(x=Zn),aes(x))+geom_histogram(aes(y =..density..),fill="#00AFBB",color="black")+
stat_function(fun = dnorm, args = list( 0, 1),aes(fill="black"),size=1.2)+theme_minimal()+
      labs(title=substitute(paste("n= ",n,", B= ",B,",a= ",parametro1,",b= ",parametro2,sep=""),
       list(n = match.call()$n,B=match.call()$B,parametro1=match.call()$a,parametro2=match.call()$b))))
}
```

```r
Sim.Exp = function(n,B,lambda){
  X  <- matrix(rexp(n*B,lambda),B,n)
  Zn <- sqrt(n)*(rowMeans(X)-(1/lambda))/sqrt(1/(lambda^2))
  (ggplot(data.frame(x=Zn),aes(x))+geom_histogram(aes(y =..density..),fill="#00AFBB",color="black")+
      stat_function(fun = dnorm, args = list( 0, 1),aes(fill="black"),size=1.2)+theme_minimal()+
      labs(title=substitute(paste("n= ",n,", B= ",B,", ",lambda,"= ",parametro,sep=""),
      list(n = match.call()$n,B=match.call()$B,parametro=match.call()$lambda))))
}
```

```r
Sim.Laplace = function(n,B,a,b){
  X  <- matrix(rlaplace(n*B,a,b),B,n)
  Zn <- sqrt(n)*(rowMeans(X)-a)/sqrt(2*(b^2))
  (ggplot(data.frame(x=Zn),aes(x))+geom_histogram(aes(y =..density..),fill="#00AFBB",color="black")+
      stat_function(fun = dnorm, args = list( 0, 1),aes(fill="black"),size=1.2)+theme_minimal()+
      labs(title=substitute(paste("n= ",n,", B= ",B,", ",mu,"= ",parametro1,", ",sigma,"= ",parametro2,sep=""),
                            list(n = match.call()$n,B=match.call()$B,parametro1=match.call()$a,parametro2=match.call()$b))))
}
```

```r
a=Sim.Laplace(50,100,1,5)
b=Sim.Laplace(100,200,1,5)
c=Sim.Laplace(300,300,1,5)
d=Sim.Laplace(800,500,1,5)
e=Sim.Laplace(50,100,0.4,0.2)
f=Sim.Laplace(100,200,0.4,0.2)
g=Sim.Laplace(300,300,0.4,0.2)
h=Sim.Laplace(800,500,0.4,0.2)
i=Sim.Laplace(50,100,40,120)
j=Sim.Laplace(100,200,40,120)
l=Sim.Laplace(300,300,40,120)
m=Sim.Laplace(800,500,40,120)
```
