```python
from plotnine import *
from scipy.stats import bernoulli,norm,expon,poisson,beta,laplace
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def SimPois(n,B,theta):
    x = poisson.rvs(theta, size=n*B).reshape(B,n)
    Zn=np.sqrt(n)*(np.mean(x,axis=1)-theta)/np.sqrt(theta)
    return((ggplot(pd.DataFrame({"x":Zn}),aes("x"))+geom_histogram(aes(y ="..density.."),
    fill="#00AFBB",color="black")+
    stat_function(fun = norm.pdf, args=dict( loc=0, scale=1),size=1.2)+
    theme_minimal()))
```

```python
def SimBern(n,B,p):
    x = bernoulli.rvs(p, size=n*B).reshape(B,n)
    Zn=np.sqrt(n)*(np.mean(x,axis=1)-p)/np.sqrt(p*(1-p))
    return((ggplot(pd.DataFrame({"x":Zn}),aes("x"))+geom_histogram(aes(y ="..density.."),
    fill="#00AFBB",color="black")+
    stat_function(fun = norm.pdf, args=dict( loc=0, scale=1),size=1.2)+
    theme_minimal()))
```

```python
def SimBeta(n,B,a,b):
    x = beta.rvs(a=a,b=b, size=n*B).reshape(B,n)
    Zn=np.sqrt(n)*(np.mean(x,axis=1)-(a/(a+b)))/np.sqrt((a*b)/(((a+b)**2)*(a+b+1)))
    return((ggplot(pd.DataFrame({"x":Zn}),aes("x"))+geom_histogram(aes(y ="..density.."),
    fill="#00AFBB",color="black")+
    stat_function(fun = norm.pdf, args=dict( loc=0, scale=1),size=1.2)+
    theme_minimal()))
```

```python
def SimExp(n,B,theta):
    x = expon.rvs(scale=1/theta, size=n*B).reshape(B,n)
    Zn=np.sqrt(n)*(np.mean(x,axis=1)-(1/theta))/np.sqrt((1/(theta**2)))
    return((ggplot(pd.DataFrame({"x":Zn}),aes("x"))+geom_histogram(aes(y ="..density.."),
    fill="#00AFBB",color="black")+
    stat_function(fun = norm.pdf, args=dict( loc=0, scale=1),size=1.2)+
    theme_minimal()))
```

```python
def SimLaplace(n,B,a,b):
    x = laplace.rvs(loc=a,scale=b, size=n*B).reshape(B,n)
    Zn=np.sqrt(n)*(np.mean(x,axis=1)-a)/np.sqrt(2*(b**2))
    return((ggplot(pd.DataFrame({"x":Zn}),aes("x"))+geom_histogram(aes(y ="..density.."),
    fill="#00AFBB",color="black")+
    stat_function(fun = norm.pdf, args=dict( loc=0, scale=1),size=1.2)+
    theme_minimal()))
```
